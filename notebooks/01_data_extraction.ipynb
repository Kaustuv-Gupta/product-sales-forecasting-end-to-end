#**Load Dataset**
---

```python
# Load the datasets
train_df = pd.read_csv('../data/experiments/raw/train.csv')
```

```python
train_df['Date'] = pd.to_datetime(train_df['Date'])
train_df = train_df.sort_values(['Date','Store_id'])
```

```python
train_df.agg(min_date=('Date', 'min'), max_date=('Date', 'max')).reset_index()
```

```python
train_df['Date'].nunique()
```

```python
train_df['Store_id'].nunique()
```

```python
unique_counts = train_df.groupby('Store_id').agg(unique_counts=('Date', 'nunique'))
unique_counts[unique_counts['unique_counts'] !=516]
```

```python
unique_counts = train_df.groupby('Store_id').agg(unique_counts=('Store_Type', 'nunique'))
unique_counts[unique_counts['unique_counts'] !=1]
```

```python
unique_counts = train_df.groupby('Store_id').agg(unique_counts=('Location_Type', 'nunique'))
unique_counts[unique_counts['unique_counts'] !=1]
```

```python
unique_counts = train_df.groupby('Store_id').agg(unique_counts=('Region_Code', 'nunique'))
unique_counts[unique_counts['unique_counts'] !=1]
```

```python
unique_counts = train_df.groupby('Date').agg(unique_counts=('Holiday', 'nunique'))
unique_counts[unique_counts['unique_counts'] !=1]
```

```python
unique_counts = train_df.groupby([ 'Date','Region_Code','Location_Type','Store_Type']).agg(unique_counts=('Discount', 'nunique'))
unique_counts[unique_counts['unique_counts'] >1]
```

**Observations:**
* **Dataset Scale**: Contains 516 daily observations across 365 unique stores.
* **Data Integrity**: Store attributes (Type, Location, and Region Code) are consistent across all records, indicating high data quality.
* **Operational Synchrony**: Holiday schedules are uniform across the entire store network.
* **Promotion Strategy**: Discounts are non-uniform and managed at the individual store level, suggesting a localized pricing strategy.
